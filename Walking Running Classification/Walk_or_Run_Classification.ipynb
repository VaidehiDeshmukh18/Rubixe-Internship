## Model Building

```python
X=data.drop(['date','time','username','activity'],axis=1)
y=data['activity']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

#### Model buliding using Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor()
rf_reg.fit(X_train, y_train)
```

```python
y_test
```

```python
y_predict= rf_reg.predict(X_test)
y_predict
```

#### Model buliding using SVM

```python
from sklearn.svm import SVC
```

```python
svm_classifier=SVC()
svm_classifier.fit(X,y)
```

```python
pred=svm_classifier.predict(X_test)
```

```python
pred
```

#### Metrics for Random Forest 

```python
r2 = r2_score(y_test, y_predict)
mse = mean_squared_error(y_test, y_predict)
mae = mean_absolute_error(y_test, y_predict)
print('r2_score: ',r2)
print('mean_squared_error: ',mse)
print('mean_absolute_error: ',mae)
print('root_mean_squared_error: ', np.sqrt(mean_squared_error(y_test, y_predict)))
```

#### Metrics for SVM

```python
r2 = r2_score(y_test, pred)
mse = mean_squared_error(y_test, pred)
mae = mean_absolute_error(y_test, pred)
print('r2_score: ',r2)
print('mean_squared_error: ',mse)
print('mean_absolute_error: ',mae)
print('root_mean_squared_error: ', np.sqrt(mean_squared_error(y_test, pred)))
```
